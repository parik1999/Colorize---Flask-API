Client side request

```python
import requests

with open('bw_image.jpg','rb') as fp:
    content = fp.read()

r = requests.get('http://662c7deeed6a.ngrok.io/img.jpeg', data=content)

with open("color_image.jpg", "wb") as fp:
    fp.write(r.content)
```

Colored image returned from server.

```python
import matplotlib.pyplot as plt
from PIL import Image
im1 = Image.open('bw_image.jpg')
im2 = Image.open('color_image.jpg')
f, axarr = plt.subplots(1,2)
axarr[0].imshow(im1)
axarr[1].imshow(im2)
```
